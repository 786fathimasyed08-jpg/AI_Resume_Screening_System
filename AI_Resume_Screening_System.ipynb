```python
import pdfplumber
import re
import nltk

from nltk.corpus import stopwords
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

nltk.download('stopwords')

print("Libraries imported successfully!")
```

```python
from google.colab import files

uploaded = files.upload()
```

```python
resume_file = list(uploaded.keys())[0]

with pdfplumber.open(resume_file) as pdf:
    resume_text = ""

    for page in pdf.pages:
        text = page.extract_text()
        if text:
            resume_text += text + "\n"

print(resume_text)
```

```python
import string

# Convert to lowercase
clean_text = resume_text.lower()

# Remove punctuation
clean_text = clean_text.translate(str.maketrans('', '', string.punctuation))

# Remove stopwords
stop_words = set(stopwords.words('english'))

words = clean_text.split()

filtered_words = [word for word in words if word not in stop_words]

clean_text = " ".join(filtered_words)

print(clean_text)
```

```python
skills = [
    "python","java","c","c++","sql","html","css","javascript",
    "react","node.js","flask","django",
    "machine learning","deep learning","nlp",
    "computer vision","tensorflow","keras","pytorch",
    "scikit-learn","numpy","pandas","opencv",
    "mysql","mongodb","firebase",
    "aws","azure","gcp",
    "git","github",
    "excel","power bi","tableau"
]

found_skills = []

for skill in skills:
    if skill in clean_text:
        found_skills.append(skill)

print("Skills found in Resume:")
print(found_skills)
```

```python
job_description = """
We are looking for an AI/ML Engineer.

Required Skills:
Python
Machine Learning
Deep Learning
NLP
Scikit-learn
TensorFlow
Git
GitHub
SQL
Flask
Pandas
NumPy
"""
```

```python
vectorizer = TfidfVectorizer()

vectors = vectorizer.fit_transform([clean_text, job_description.lower()])

similarity = cosine_similarity(vectors)[0][1]

match_percentage = round(similarity * 100, 2)

print("Resume Match Percentage:", match_percentage, "%")
```

```python
job_skills = [
    "python",
    "machine learning",
    "deep learning",
    "nlp",
    "scikit-learn",
    "tensorflow",
    "git",
    "github",
    "sql",
    "flask",
    "pandas",
    "numpy"
]

missing_skills = []

for skill in job_skills:
    if skill not in found_skills:
        missing_skills.append(skill)

print("Missing Skills:")
print(missing_skills)
```

```python
print("\n----- Resume Suggestions -----")

if len(missing_skills) == 0:
    print("✅ Excellent! Your resume matches the job requirements.")
else:
    print("You should consider learning or adding these skills:")
    for skill in missing_skills:
        print("•", skill)

print("\nKeep your GitHub updated.")
print("Add certifications and relevant projects.")
print("Customize your resume for each job.")
```

```python
ats_score = 100 - (len(missing_skills) * 10)

if ats_score < 0:
    ats_score = 0

print("ATS Score:", ats_score, "/100")
```

```python
print("="*50)
print("      AI RESUME SCREENING REPORT")
print("="*50)

print("\nResume Match Percentage:", match_percentage, "%")
print("ATS Score:", ats_score, "/100")

print("\nSkills Found:")
for skill in found_skills:
    print("✅", skill)

print("\nMissing Skills:")
for skill in missing_skills:
    print("❌", skill)

print("\nSuggestions:")
if len(missing_skills) == 0:
    print("Excellent! Your resume is a good match.")
else:
    print("Add these skills to improve your resume:")
    for skill in missing_skills:
        print("➡", skill)
```

```python
report = f"""
AI Resume Screening Report

Resume Match Percentage: {match_percentage} %

ATS Score: {ats_score}/100

Skills Found:
{found_skills}

Missing Skills:
{missing_skills}
"""

with open("Resume_Report.txt", "w") as file:
    file.write(report)

print("Resume report saved successfully!")
```

```python
if ats_score >= 90:
    rating = "★★★★★ Excellent"
elif ats_score >= 75:
    rating = "★★★★ Very Good"
elif ats_score >= 60:
    rating = "★★★ Good"
elif ats_score >= 40:
    rating = "★★ Average"
else:
    rating = "★ Needs Improvement"

print("Overall Resume Rating:", rating)
```
